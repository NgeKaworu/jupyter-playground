```python
import matplotlib.pyplot as plt
import numpy as np

# FV = PV * (1 + r)^n
# PV = FV / (1 + r)^n
# r = (FV / PV)^(1/n) - 1
# n = ln(FV / PV) / ln(1 + r)

def variablePV():
    count = np.arange(10, 21, 1)
    PV = count
    r = 0.2
    n = 10
    FV = PV * np.power((1 + r), n)
    
    plt.plot(count,    PV, count,  FV)
       
    # 设置x轴和y轴的刻度
    plt.xticks(np.arange(10, 20, 1))
    plt.yticks(np.arange(10, max(FV)+5, 5))

    # 在每个点旁边显示数值
    for i in range(len(count)):
        plt.text(count[i], PV[i], f"{PV[i]:.2f}", ha='center', va='bottom')
        plt.text(count[i], FV[i], f"{FV[i]:.2f}", ha='center', va='bottom')
    
    # 添加图例
    plt.legend()
    
    plt.show()

variablePV()
```
